```python
import matplotlib.pyplot as plt
import numpy as np



# Calculate total positive and negative predictions
positive_predictions = 564 + 2  # False Positives + True Positives
negative_predictions = 3 + 67388  # False Negatives + True Negatives

# Create figure with two subplots side by side
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))  # Reduced width from 16 to 12

# First subplot - Model Predictions
categories1 = ['Positive Predictions', 'Negative Predictions']
values1 = [positive_predictions, negative_predictions]

bars1 = ax1.bar(categories1, values1, color=['#2ecc71', '#3498db'], edgecolor='white', width=0.4)  # Reduced width from 0.6 to 0.4

# Adding value labels for first subplot
for bar in bars1:
    height = bar.get_height()
    ax1.text(bar.get_x() + bar.get_width()/2, height, 
            f'{int(height):,}',
            ha='center', va='bottom',
            fontsize=10)

# Customize first subplot
ax1.set_title('AR Model Predictions Distribution in PC3', 
             fontsize=16, fontweight='bold', pad=20)
ax1.set_ylabel('Count', fontsize=12, labelpad=10)
ax1.spines['top'].set_visible(False)
ax1.spines['right'].set_visible(False)
ax1.yaxis.grid(True, linestyle='--', alpha=0.3)
ax1.set_axisbelow(True)
ax1.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: format(int(x), ',')))

# Second subplot - TOBIAS analysis
categories2 = ['AR Bound', 'AR Unbound']
values2 = [1970, 67957 - 1970]

bars2 = ax2.bar(categories2, values2, color=['#2ecc71', '#3498db'], edgecolor='white', width=0.4)  # Reduced width from 0.6 to 0.4

# Adding value labels for second subplot
for bar in bars2:
    height = bar.get_height()
    ax2.text(bar.get_x() + bar.get_width()/2, height, 
            f'{int(height):,}',
            ha='center', va='bottom',
            fontsize=10)

# Customize second subplot
ax2.set_title('TOBIAS AR Binding Distribution in PC3', 
             fontsize=16, fontweight='bold', pad=20)
ax2.set_ylabel('Count', fontsize=12, labelpad=10)
ax2.spines['top'].set_visible(False)
ax2.spines['right'].set_visible(False)
ax2.yaxis.grid(True, linestyle='--', alpha=0.3)
ax2.set_axisbelow(True)
ax2.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: format(int(x), ',')))

# Adjust layout
plt.tight_layout()

# Display the plot
plt.show()
```
