```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data = {
    'cell_line_added': ['Ishikawa', 'AN3_CA', 'LNCAP', 'HEK_293', 'C4-2', 'A549', 'RT4', 'HuH-7'],
    'accuracy': [78.4676, 79.3459, 89.7013, 90.2502, 90.0496, 91.8496, 88.8363, 90.0382],
}

df = pd.DataFrame(data)

sns.set(style='whitegrid')
plt.figure(figsize=(12, 6))

# Create a line plot with markers
ax = sns.lineplot(x='cell_line_added', y='accuracy', data=df, marker='o')

# Rotate x-axis labels
plt.xticks(rotation=45)

# Add labels to each point
for i, row in df.iterrows():
    ax.text(i, row['accuracy'] + 0.3, f'{row["accuracy"]:.2f}', ha='center')

plt.title('Accuracy of Cell Lines on 22Rv1')
plt.xlabel('Cell Line Added')
plt.ylabel('Accuracy')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data = {
    'Number of Training Chromosomes': [3, 6, 10, 15, 22],
    'accuracy': [86.6584, 87.0246, 90.3289, 90.7358, 90.2133],
}

df = pd.DataFrame(data)

sns.set(style='whitegrid')
plt.figure(figsize=(12, 6))

# Create a line plot with markers
ax = sns.lineplot(
    x='Number of Training Chromosomes',
    y='accuracy',
    data=df,
    marker='o',
    linewidth=2
)

# Set x-axis ticks to the provided values
ax.set_xticks(df['Number of Training Chromosomes'])

# Add labels to each point
for x, y in zip(df['Number of Training Chromosomes'], df['accuracy']):
    ax.text(x, y + 0.1, f'{y:.2f}', ha='center')

plt.title('Accuracy on Chromosomes 7 and 11')
plt.xlabel('Number of Training Chromosomes')
plt.ylabel('Accuracy')

plt.tight_layout()
plt.show()
```
